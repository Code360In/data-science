```
# df = spark2.read.csv('appl_stock.csv',inferSchema=True,header=True)
df = sqlContext.sql('SELECT * FROM appl_stock_csv')
```

```
df.select(['Date','Open']).show()
```

```
from pyspark.sql.functions import (dayofmonth
                                   ,hour
                                   ,dayofyear
                                   ,month
                                   ,year
                                   ,weekofyear
                                   ,format_number
                                   ,date_format)
```

```
df.select(dayofmonth(df['Date'])).show()
```

```
df.select(hour(df['Date'])).show()
```

```
# df.select(year(df['Date']))
dfAAPL = df.withColumn('Year',year(df['Date']))
```

```
dfAAPL.groupBy('Year').mean().show()
```

```
resultAAPL = dfAAPL.groupBy('Year').mean().select(['Year','avg(Close)'])
```

```
resultAAPL.show()
```

```
resultNEW = resultAAPL.withColumnRenamed('avg(Close)','Average Closing Price')
```

```
resultNEW.select(['Year',format_number('Average Closing Price',2).alias('Average Close')]).show()
```
